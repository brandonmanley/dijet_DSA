```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator
import dijet_utils as dju
%matplotlib inline
```

```python
# import data file

mcfile = 'example_data.npy'
mc_df = pd.DataFrame(np.load(mcfile), columns=['s', 'Q', 'x', 'delta', 'pT', 'z', 'y', 'phi_kp', 'phi_Dp', 
                                      'num_dsa', 'den_dsa', '<1>', '<cos(phi_kp)>', '<cos(phi_Dp)>', 
                                     '<cos(phi_Dp)cos(phi_kp)>', '<sin(phi_Dp)sin(phi_kp)>'])

# set up plotting conventions
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
plt.rcParams["legend.fontsize"] = 12
plt.rcParams["axes.labelsize"] = 14 
plt.rcParams["xtick.labelsize"] = 12  
plt.rcParams["ytick.labelsize"] = 12
plt.rcParams["axes.titlesize"] = 16  
plt.rcParams["text.usetex"] = True
plt.rcParams["text.latex.preamble"] = r"\usepackage{amsmath}"

mc_df.tail()
```

```python
# example plots (see README for plot_histogram documentation)

# plot numerator and denominator of asymmetry as a function of Q 
dju.plot_histogram([mc_df], 'Q', ['den_dsa', 'num_dsa'], 
                   {}, 
                   lumi = 10,
                   y_scale='log', 
                   nbins = 20, 
                   min_pT_line = True,
                   y_label=r'$d\sigma /dQ \,\,[\mathrm{fb}*\mathrm{GeV}^{-1}]$', 
                   x_label=r'$Q\, [\mathrm{GeV}]$'
              )


# plot correlations of asymmetry as a function of p_T 
dju.plot_histogram([mc_df], 'pT', ['<cos(phi_Dp)cos(phi_kp)>', '<sin(phi_Dp)sin(phi_kp)>','<cos(phi_Dp)>'], 
                   {
                    'Q': [7.5, 8],
                    'z': [0.3, 0.4]
                   }, 
                   lumi = 10,
                   efficiency = 1, 
                   nbins = 20, 
                   y_limits = [-0.03,0.04],
                   zero_line = True,
                   min_pT_line = True,
                   y_label=r'$d \langle f(\phi_{kp}, \phi_{\Delta p}) \rangle /dp_T \,\,[\mathrm{GeV}^{-1}]$', 
                   x_label=r'$p_T\, [\mathrm{GeV}]$'
                  )

```
