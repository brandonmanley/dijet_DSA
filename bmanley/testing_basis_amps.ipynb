```python
data_file = 'narr_ymin4.61_ymax14.91_AAMS09.dat'
data = np.loadtxt(data_file)

asp_ratio = 4/3
psize = 5
fig, ax = plt.subplots(1,1, figsize=(asp_ratio*psize*1, psize*1))

delta = 0.01


logr_values = np.arange(-13.8, len(data[0])*delta - 13.8, delta)

colors = ['red', 'blue', 'green']
for iplot, iy in enumerate([5, 10, 14]):
    y_index = round((1/delta)*(iy - 4.61))
    ax.scatter(logr_values[800::], data[y_index][800::], label=rf'$\ln W^2 / \Lambda^2  = {iy}$', color=colors[iplot])

ax.set_title(r'$N \left( \ln r, \ln \frac{W^2}{\Lambda^2} \right)$')
# ax.semilogx()
ax.set_xlabel(r'$\ln r$')
ax.legend(frameon=False)
plt.show()

```
